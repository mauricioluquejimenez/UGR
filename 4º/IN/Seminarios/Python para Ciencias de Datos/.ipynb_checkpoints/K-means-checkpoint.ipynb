```python
import time
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import datetime
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.impute import KNNImputer
from math import floor
from sklearn.preprocessing import LabelEncoder
from yellowbrick.cluster import KElbowVisualizer
```

#### Normalizacion de los datos

```python
def norm_to_zero_one(df):
    return (df - df.min()) * 1.0 / (df.max() - df.min())
```

#### Reorganizacion de datos

```python
datos["Age"] = 2025 - datos["Year_Birth"]
```

```python
datos["Education"].value_counts()
```

```python
datos["Marital_Status"].value_counts()
```

```python
datos["Marital_status"] = datos["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Single":"Alone", "Divorced":"Alone", "Widow":"Alone", "Absurd":"Alone", "YOLO":"Alone",})
datos["Marital_status"].value_counts()
```

```python
datos["Children"] = datos["Kidhome"] + datos["Teenhome"]
```

```python
datos["Family_Size"] = datos["Marital_status"].map({"Alone":1, "Partner":2}) + datos["Children"]
```

```python
datos["Is_Parent"] = np.where(datos.Children>0,1,0)
```

```python
datos["Dt_Customer"] = pd.to_datetime(datos["Dt_Customer"].astype(str), errors="coerce", dayfirst=True)
fecha_mas_nueva = datos ["Dt_Customer"].max()
fecha_mas_antigua = datos ["Dt_Customer"].min()

print(f"Fecha de inscripcion del cliente mas reciente registrada: {fecha_mas_nueva.date()}")
print(f"Fecha de inscripcion del cliente mas antigua registrada: {fecha_mas_antigua.date()}")
```

```python
datos["Total_Spent"] = datos["MntWines"] +datos["MntFruits"] +datos["MntMeatProducts"] +datos["MntFishProducts"] + datos["MntSweetProducts"] +datos["MntGoldProds"]
```

```python
datos = datos.rename(columns={"MntWines":"Wines", "MntFruits":"Fruits", "MntMeatProducts":"Meat", "MntFishProducts":"Fish", "MntSweetProducts":"Sweet", "MntGoldProds":"Gold"})
```

```python
eliminar_columnas =["Marital_Status", "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth"]
datos = datos.drop(eliminar_columnas, axis=1)
```

#### One hot encoding

```python
le = LabelEncoder()
categorical_cols = ["Education", "Marital_status"]
for col in categorical_cols:
    datos[col] = le.fit_transform(datos[col])
```

```python
datos.columns.to_list()
```

```python
variables_usadas = ['Income','Age', 'Children', 'Family_Size', 'Is_Parent', 'Total_Spent']
```

```python
X = datos[variables_usadas]
print(X)
```

#### Normalizar

```python
X_normal = X.apply(norm_to_zero_one) 
```

```python
X_normal.head()
```

#### K-Means

```python
kmeans = KMeans(random_state=12345)
visualizer = KElbowVisualizer(kmeans, k=(1,11))
visualizer.fit(X_normal)
visualizer.show()
```

```python
print(" ----Ejecutando k-Means",end="")
k_means = KMeans(init="k-means++", n_clusters=2, n_init=5, random_state=12345)
t = time.time()
cluster_predict = k_means.fit_predict(X_normal)
tiempo = time.time() - t
print(": {:.5f} segundos. ".format(tiempo), end="")
```

#### Indice Calinski-Harabasz

```python
metric_CH = metrics.calinski_harabasz_score(X_normal, cluster_predict)
print("Indice Calinski-Harabasz: {:.5f}".format(metric_CH))
```

#### Silhouette Coefficient

```python
metric_SC = metrics.silhouette_score(X_normal, cluster_predict, metric="euclidean")
print("Silhouette Coefficient: {:.5f}".format(metric_SC))
```

#### Tamaño de los clusters

```python
clusters = pd.DataFrame(cluster_predict, index=X.index, columns=["cluster"])

print("Tamaño de cada cluster:")
size = clusters["cluster"].value_counts().sort_index()

for i, c in size.items():
    print("%d: %5d (%5.2f%%)" % (i+1, c, 100 * c / len(clusters)))
```
